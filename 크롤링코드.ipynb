```python
import os
import json
import requests
import random
import time
from django.core.files.base import ContentFile
from food.models import Category, Shop, Item
from urllib.parse import urljoin
```

```python
class Food_date:
    HOST_URL = 'https://www.yogiyo.co.kr'
    
    def __init__(self, headers):
        self.headers = headers
        
    def get_restaurant_list_by_geo(self, zip_code):
        url = self.HOST_URL + '/api/v1/restaurants-geo/'
        params = dict(
            item=20,
            order='rank',
            page=0,
            search='',
            zip_code=zip_code,
        )
        res = requests.get(url, params=params, headers=self.headers)
        return res.json()
    
    def get_menu_list(self, restaurant_id):
        base_url = self.HOST_URL + '/api/v1/restaurants/{restaurant_id}/menu/?add_photo_menu=original'
        url = base_url.format(restaurant_id=restaurant_id)
        res = requests.get(url, headers=self.headers)
        return res.json()
```
